#Visualization

```
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

G=nx.from_pandas_edgelist(data1, 'FromNodeId', 'ToNodeId')
```

```
plt.figure(figsize=(10,10))
plt.title('Meeting of 18 women across 14 events')
nx.draw(G, with_labels=True)
#nx.draw(G)
plt.savefig('C1_W3_Gautham.png')
```

The data is called Southern Women dataset collected in 1941 observing the social activities of 18 women over 14 events across a period of 9 months. The visualization shows an undirected and acyclic graph where the women organize themselves in around 2 distinct groups and meet at these events. Moreover, these groups seems to have core and peripheral members. Despite the data being small, it is considered to be a touchstone for comparing analytical methods in social network analysis. The dataset has been analysed by various researchers over the years and have found that it can also be complicated while examining the pattern of the edges which seems to be frayed.
