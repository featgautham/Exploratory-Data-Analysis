```
import pandas as pd
data1 = pd.read_csv('ExcelFormattedGISTEMPData2CSV.csv')
print(data1.shape)
data1.head()
```

#Visualization

```
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.plot('Year', 'Glob', data=data1, color='black', label='Global Temperature')
plt.plot('Year', 'NHem', data=data1, color='brown', label='North Hemisphere Temperature')
plt.plot('Year', 'SHem', data=data1, color='orange', label='South Hemisphere Temperature')
plt.xlabel('Years')
plt.ylabel('Deviation of Temperatures (in celsius)')
plt.yticks(list(range(-100,100,10)))
plt.title('Global and hemispheric temperatures over the years')
plt.legend()

plt.savefig('plot.png')
```

A subset of GISTEMP data which is the years, global, northern and southern hemispheric temperatures is considered for this visualization where the deviation of global and hemispheric temperatures is represented in y axis and its years in x axis. The visualization clearly shows that temperatures across the globe and hemispheres are on the rise over the years. The colors used to represent the temperatures are saturated and contrasting to each other, helping to differentiate from each other for easy interpretation.
